```python
import matplotlib.pyplot as plt
file = open("maps.txt")
mAP = []
count = 1
ten_map = []
total = 0
for line in file.readlines():
    index = line.index(",")
    current_map = float(line[4:index])
    mAP.append(current_map)
    count += 1
    
    if count%10 != 0 :
        total += current_map
    else:
        ten_map.append(total/10)
        total = 0
        
        
plt.plot(mAP[150:1500])
plt.show()
plt.plot(ten_map[15:150])
plt.show()
```
